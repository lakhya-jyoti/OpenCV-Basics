**Histogram Equalization** a graphical representation of the intensity distribution of an image.

It quantifies the number of pixels for each intensity value considered.


```python
# Mounting the Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importing OpenCV and Matplotlib
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Loading an image from Google Drive folder
folder_path = '/content/drive/My Drive/sample-flowers'
image_file = 'flower-photo-01.jpg'
image_path = os.path.join(folder_path, image_file)
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Displaying the original image
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()

# Converting the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)  # Change here

# Displaying the grayscale image
plt.imshow(gray_image, cmap='gray')  # Specify the colormap
plt.title('Original Grayscale Image')
plt.axis('off')
plt.show()

# Applying histogram equalization
equalized_image = cv2.equalizeHist(gray_image)

# Displaying the equalized image
plt.imshow(equalized_image, cmap='gray')  # Specify the colormap
plt.title('Equalized Image')
plt.axis('off')
plt.show()

# Displaying the histogram of the original and equalized images
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

axs[0].hist(gray_image.ravel(), bins=256, color='blue', alpha=0.7)
axs[0].set_title('Histogram of Original Grayscale Image')

axs[1].hist(equalized_image.ravel(), bins=256, color='red', alpha=0.7)
axs[1].set_title('Histogram of Equalized Grayscale Image')

plt.tight_layout()
plt.show()
```
