# **Open CV Arithmetic Operations**

**Note:** Store the images on Google Drive.
The flower image is uploaded you can download this too.

```python
# Mounting Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importing OpenCV, OS, NumPy, and Matplotlib
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Loading images from Google Drive folder
folder_path = '/content/drive/My Drive/sample-flowers'
image_file1 = 'flower-photo-01.jpg'
image_file2 = 'flower-photo-02.jpg'
image_path1 = os.path.join(folder_path, image_file1)
image_path2 = os.path.join(folder_path, image_file2)

image1 = cv2.imread(image_path1)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.imread(image_path2)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

# Displaying the images using subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(image1)
axes[0].axis('off')
axes[0].set_title('Flower Photo 01')

axes[1].imshow(image2)
axes[1].axis('off')
axes[1].set_title('Flower Photo 02')

plt.show()
```

# **Adding**

both images should be of equal size and depth.

```python
new_added_img = cv2.add(image1, image2)
plt.imshow(new_added_img)
plt.title('Added Image')
plt.axis('off')
plt.show()
```

# **Subtraction**
The images should be of equal size and depth.

```python
subtracted = cv2.subtract(image1, image2)
plt.imshow(subtracted)
plt.title('Subtracted Image')
plt.axis('off')
plt.show()
```

# **Blending**

```python
Blend = cv2.addWeighted(image1,0.7,image2,0.3,0)
plt.imshow(Blend)
plt.title('Blended Image')
plt.axis('off')
plt.show()
```

# **Bitwise Operation**

```python
# Perform bitwise AND operation
bitwise_and = cv2.bitwise_and(image1, image2)
plt.imshow(bitwise_and)
plt.title('Bitwise AND')
plt.axis('off')
plt.show()

# Perform bitwise OR operation
bitwise_or = cv2.bitwise_or(image1, image2)
plt.imshow(bitwise_or)
plt.title('Bitwise OR')
plt.axis('off')
plt.show()

# Perform bitwise XOR operation
bitwise_xor = cv2.bitwise_xor(image1, image2)
plt.imshow(bitwise_xor)
plt.title('Bitwise XOR')
plt.axis('off')
plt.show()

# Perform bitwise NOT operation on image1
bitwise_not_image1 = cv2.bitwise_not(image1)
plt.imshow(bitwise_not_image1)
plt.title('Bitwise NOT (Image 1)')
plt.axis('off')
plt.show()
```
